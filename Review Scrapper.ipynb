```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.flipkart.com/nothing-phone-1-black-128-gb/p/itmeea53a564de47')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

# Fetching Reviews

```python
#fetching the reviews
reviews = soup.findAll('div',{'class':'t-ZTKy'});
for review in reviews:
    print(review.get_text() + "\n\n" )
```

# Fetching Ratings

```python
#Overall Rating
ratings = soup.find('div',{'class' : '_2d4LTz'}).get_text();
print(ratings)
```

```python
#Individual Ratings
individual_ratings = soup.findAll('div',{'class' : '_3LWZlK _1BLPMq'});
for indivi_rating in individual_ratings:
    print(indivi_rating.get_text() + " \n")
```

# Fetching Customer Name

```python
customer_name = soup.findAll('p',{'class' :'_2sc7ZR _2V5EHH'});
for custom_name in customer_name:
    print(custom_name.get_text() + " \n")
```

# Fetching Price

```python
#fetching price

price = soup.find('div',{'class':'_30jeq3 _16Jk6d'}).get_text();
print(price)
```
